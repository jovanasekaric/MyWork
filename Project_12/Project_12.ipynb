```python
def clean_contractions(text, mapping):
    specials = ["", "", "", "`"]
    for s in specials:
        text = text.replace(s, "'")
    text = ' '.join([mapping[t] if t in mapping else t for t in text.split(" ")])
    return text
```

### Normalization

- Lemmatization - is a text normalization technique used in Natural Language Processing (NLP), that switches any kind of a word to its base root mode. Lemmatization is responsible for grouping different inflected forms of words into the root form, having the same meaning

```python
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatization(text):
    lemmatized_sentence = " ".join(lemmatizer.lemmatize(token) for token in text)
    return lemmatized_sentence
```

```python
%%time
data["lemmatizer_review"] = data['tokens'].apply(lemmatization)
```

- Stemming -  is the process of reducing a word to its stem that affixes to suffixes and prefixes or to the roots of words known as "lemmas". Stemming is important in natural language understanding (NLU) and natural language processing (NLP).

```python
#Stemming the text

ps = nltk.porter.PorterStemmer()


def stemmer(text):
    text = ' '.join([ps.stem(token) for token in text])
    return text
```

```python
%%time
data['stemmer_review'] = data['tokens'].apply(stemmer)
```

#### Cross validation

Now, I will define function for CV.Three parameters are passed to the function: model, training data, test data. The output is mean accuracy of CV scores.

```python
def CV(model, X, Y):
    fitted = model.fit(X, Y)
    accuracy = cross_val_score(fitted, X, Y, cv=5,scoring='accuracy')
    print(accuracy.mean())
```

```python
CV(LinearSVC(), X_train, y_train)
CV(MultinomialNB(), X_train, y_train)
CV(SGDClassifier(), X_train, y_train)
```

#### Import packages

```python
from tensorflow import keras
import gensim.downloader as api
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

Split the data into training and testing sets:

```python
X_train, X_test, y_train, y_test = train_test_split(data['review'], data['sentiment'], test_size=0.2, random_state=42)
```

Convert the text data into numerical sequences using tokenization:

```python
tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(X_train)
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)
```

Pad the numerical sequences to ensure that all sequences have the same length:

```python
max_len = 200
X_train_padded = pad_sequences(X_train_seq, maxlen=max_len, padding='post', truncating='post')
X_test_padded = pad_sequences(X_test_seq, maxlen=max_len, padding='post', truncating='post')
```

```python
size_of_vocabulary = len(tokenizer.word_index) + 1
print(size_of_vocabulary)
```

Train a Word2Vec model (neural network-based technique used for generating word embeddings, which are vector representations of words in a high-dimensional space) on the training data:

```python
sentences = list(X_train)
tokenized_texts = [word_tokenize(text) for text in sentences]
print([' '.join(row) for row in tokenized_texts[:2]])
```

```python
word2vec_model = Word2Vec(tokenized_texts,
                 vector_size=100,      # embedding vector size
                 min_count=5,  # consider words that occured at least 5 times
                 window=5)  # define context as a 5-word window around the target word

embedding_dim = 100
```

Get the word index and embedding matrix (to represent words in a numerical format that can be easily processed by machine learning algorithms):

```python
word_index = tokenizer.word_index
num_words = len(word_index) + 1
embedding_matrix = np.zeros((num_words, embedding_dim))
for word, i in word_index.items():
    if word in word2vec_model.wv:
        embedding_matrix[i] = word2vec_model.wv[word]
```

### Methods

Now, I will define methods to make my code more readable:

```python
def fit_model(model, model_path):

    # Set up early stopping and model checkpoint callbacks
    early_stopping_callback = EarlyStopping(monitor='val_loss', patience=3)
    model_checkpoint_callback = ModelCheckpoint(filepath=model_path, monitor='val_loss', save_best_only=True)

    # Train the model on the training data with early stopping and model checkpoint callbacks
    history = model.fit(X_train_padded, y_train, epochs=5, batch_size=32, validation_split=0.1, callbacks=[early_stopping_callback, model_checkpoint_callback])
    return history
```

```python
def evaluate_model(model, model_path, results, model_name):

    # Load the best model saved by the model checkpoint callback
    model.load_weights(model_path)

    # Evaluate the performance of the trained model on the testing data
    loss, accuracy = model.evaluate(X_test_padded, y_test)
    print(f'Test Loss: {loss:.4f}')
    print(f'Test Accuracy: {accuracy:.4f}')

    results.append((model_name, loss, accuracy))
```

```python
def plot_confusion_matrix(model):
    y_pred = model.predict(X_test_padded)
    y_pred = np.round(y_pred)
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
def plot_learning_curve(history):
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Learning Curve')
    plt.legend()
    plt.show()
```

```python
def plot_loss_curve(history):
    plt.plot(history.history['loss'], label='Training loss')
    plt.plot(history.history['val_loss'], label='Validation loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Learning Curve')
    plt.legend()
    plt.show()
```

### A Simple LSTM Network


I will now define several different versions of the LSTM neural network and apply them to the prepared data. After that I will compare the performance.

Initiallize model:

```python
lstm_model = keras.Sequential()
lstm_model.add(keras.layers.Embedding(num_words, embedding_dim, input_length=max_len))
lstm_model.add(keras.layers.LSTM(100))
lstm_model.add(keras.layers.Dense(1, activation='sigmoid'))
lstm_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
print(lstm_model.summary())
```

Train the model on the training data:

```python
lstm_history = fit_model(lstm_model, 'best_model_lstm.h5')
```

Evaluate the performance of the trained model on the testing data:

```python
evaluate_model(lstm_model, 'best_model_lstm.h5', results, "A Simple LSTM Network")
```

Plot a confusion matrix:

```python
plot_confusion_matrix(lstm_model)
```

Plot learning curves:

```python
plot_learning_curve(lstm_history)
```

```python
plot_loss_curve(lstm_history)
```

From the graph above, I can see that the trend for accuracy is constantly increasing. In the third epoch, the accuracy of the current data increases significantly. From the third epoch, the accuracy of the validation data drops compared to the training data, which can potentially indicate overfitting.

## Hyperparameters tuning

In this section I will demonstrate how to perform Hyperparameters tuning in such models. Due to the lack of resources and time, I did not do Hyperparameters tuning for each of the models above, but it is suggested to do it. Define the model as a function that takes in the hyperparameters as arguments:

```python
def create_model(lstm_units=32, dropout=0.2):
    model = keras.Sequential([
        keras.layers.Embedding(input_dim=num_words, output_dim=embedding_dim, weights=[embedding_matrix], input_length=max_len, trainable=False),
        keras.layers.Bidirectional(keras.layers.LSTM(units=lstm_units, dropout=dropout, recurrent_dropout=dropout, return_sequences=True)),
        keras.layers.Bidirectional(keras.layers.LSTM(units=lstm_units, dropout=dropout, recurrent_dropout=dropout)),
        keras.layers.Dense(units=1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

Perform hyperparameter tuning using grid search

```python
from sklearn.model_selection import GridSearchCV
```

Define the hyperparameters to tune the model:

```python
param_grid = {
    'lstm_units': [32, 64],
    'dropout': [0.2, 0.3]
}
```

Define a GridSearchCV object to search for the best hyperparameters using cross-validation:

```python
grid_search = GridSearchCV(estimator=keras.wrappers.scikit_learn.KerasClassifier(build_fn=create_model, verbose=0),
                           param_grid=param_grid, cv=3)
```

Fit the GridSearchCV object on the training data:

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1,patience=3)
mc = ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', save_best_only=True,verbose=1)

grid_search.fit(X_train_padded,
                y_train,
                epochs=2,
                batch_size=32,
                validation_split=0.1,
                callbacks=[es, mc])
```
